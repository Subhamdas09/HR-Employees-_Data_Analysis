```python
#what is earliest and Latest date of hire for employees in each manager's team?
```

```python
# Ensure HireDate is datetime
df['DateofHire'] = pd.to_datetime(df['DateofHire'])

result = (
    df.groupby('Manager')
      .agg(Earliest_Hire=('DateofHire', 'min'),
           Latest_Hire=('DateofHire', 'max'))
      .reset_index()
)

print(result)
```
